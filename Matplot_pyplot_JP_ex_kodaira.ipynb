````
import requests
import pandas as pd
import io

url='http://tom2rd.sakura.ne.jp/hanakotech/parkmap/jinko/kodaira-h31-chobetsu.csv'
r = requests.get(url)
res = requests.get(url).content
df = pd.read_csv(io.StringIO(res.decode('utf-8')),index_col=0)
# nrows=72というのは試行錯誤です
df
````

````
#seaborn の　pairplot でそれぞれの変数の関係をプロットしてみた
sns.pairplot(df)
````

````
import numpy as np
group_data = list(df['世 帯 数'])
group_names = list(df.index)
group_mean = np.mean(group_data)
print(group_mean)
````

````
import numpy as np

group_data = list(df['世 帯 数'])
group_names = list(df.index)
group_mean = np.mean(group_data)
print(group_mean)
````

````
fig, ax = plt.subplots(figsize=(8, 20))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# Add a vertical line, here we set the style in the function call
ax.axvline(group_mean, ls='--', color='r')


# Now we'll move our title up since it's getting a little cramped
ax.title.set(y=1.05)

ax.set(xlim=[0, 6100], xlabel='世帯数', ylabel='町名',
       title='小平市町名別　世帯数')

plt.show()
````

````
import numpy as np

group_data = list(df['人 口 密 度（人 ／ ㎢）'])
group_names = list(df.index)
group_mean = np.mean(group_data)
print(group_mean)
````

````
fig, ax = plt.subplots(figsize=(8, 20))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# Add a vertical line, here we set the style in the function call
ax.axvline(group_mean, ls='--', color='r')


# Now we'll move our title up since it's getting a little cramped
ax.title.set(y=1.05)

ax.set(xlim=[0, 20000], xlabel='人口密度', ylabel='町名',
       title='小平市町名別　人口密度')

plt.show()
````
