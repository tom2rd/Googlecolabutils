````
import requests
from bs4 import BeautifulSoup
import re
# Webページを取得して解析する
load_url = "https://www.city.kodaira.tokyo.jp/kurashi/076/076079.html"
load_urlo="http://keishin.ullet.com/search/city/13211/pref/13.html#pref/13/city/13211"
html = requests.get(load_urlo)
soup = BeautifulSoup(html.content, "html.parser")
#alists=soup.find_all('a')
#alists
print(soup.find_all('a', class_="company_name"))
print(soup.find_all('a',text=re.compile("P.")))
````
